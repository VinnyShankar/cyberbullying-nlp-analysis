```python
# imports
import pandas as pd
```

```python
df1 = pd.read_json("small_channels_comments.json")
df1["channel_size"] = [0]*len(df1)
df1["identity"] = [1]*len(df1)
df1.head()
```

```python
df2 = pd.read_json("large_channels_comments.json")
df2["channel_size"] = [1]*len(df2)
df2["identity"] = [1]*len(df2)
df2.head()
```

```python
df3 = pd.read_json("lgbt_channels_comments.json")
lgbt_size_df = pd.read_csv("lgbt_size.csv")
df3["channel_size"] = lgbt_size_df["size"].tolist()
df3["identity"] = [0]*len(df3)
df3.head()
```

```python
df4 = pd.concat([df1,df2,df3])
df4
```

```python
df4.reset_index(drop=True,inplace=True)
```

```python
df4.info()
```

```python
df4.to_json("all_channels_comments.json")
```
