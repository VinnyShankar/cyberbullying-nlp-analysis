```python
# Imports
import pandas as pd
```

```python
# read csv into pandas dataframe
df = pd.read_json("large_channels_comments.json")
df.head()
```

```python
# convert comments column into list
comment_list = df["comments"].tolist()
```

```python
count = 0
for x in range(0,len(comment_list)):
    count += len(comment_list[x])

count
```

```python
for x in range(0,len(comment_list)):
    comment_list[x] = list(set(comment_list[x]))
```

```python
count = 0
for x in range(0,len(comment_list)):
    count += len(comment_list[x])

count
```

```python
for x in range(0,len(comment_list)):
    for y in range(0,len(comment_list[x])):
        if 'https' in comment_list[x][y]:
            print(comment_list[x][y])
```
