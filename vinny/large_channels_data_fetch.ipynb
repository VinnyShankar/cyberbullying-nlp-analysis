```python
# Imports
import pandas as pd
from googleapiclient.discovery import build
from config import api_key
```

```python
df = pd.read_csv("large_channels.csv")
```

```python
df.head()
```

```python
video_ids = df["video_id"].tolist()
video_ids[:5]
```

```python
comments = []
```

```python
# Counter for progress display
counter = 1

# Iterate through video ids
for id in video_ids:

    # Try
    try:

        # empty list for storing reply
        replies = []

        # creating youtube resource object
        youtube = build('youtube', 'v3',
                        developerKey=api_key)

        # retrieve youtube video results
        video_response=youtube.commentThreads().list(
        part='snippet',
        videoId=id
        ).execute()


        # iterate video response
        while video_response:

            # extracting required info
            # from each result object
            for item in video_response['items']:

                # Extracting comments
                comment = item['snippet']['topLevelComment']['snippet']['textOriginal']

                # Store reply is list
                replies.append(comment)

                # print comment with list of reply
                # print(comment, end = '\n\n')

            # Again repeat
            if 'nextPageToken' in video_response and len(replies)<1000:
                video_response = youtube.commentThreads().list(
                        part = 'snippet',
                        videoId = id,
                        pageToken = video_response['nextPageToken']
                    ).execute()
            else:
                comments.append(replies)
                break
    
    # Exception handling
    except Exception as e:
        print(e)
        comments.append(replies)
        pass

    # Counter for display
    print(f"{counter}: Done")
    counter += 1
```

```python
df["comments"] = comments
df.head()
```

```python
df.to_csv("large_channels_comments.csv",index=False)
df.to_json("large_channels_comments.json")
```
