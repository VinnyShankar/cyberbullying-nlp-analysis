```python
import csv
import pandas as pd
from googletrans import Translator, constants
# init the Google API translator
translator = Translator()

# from langdetect import detect
# import langid
# from google_trans_new import google_translator
```

```python
#read csv into pandas dataframe
df = pd.read_json("../vinny/small_channels_comments.json")
df.head()
```

```python
comment_list = df["comments"].tolist()
comment_list[0]
```

```python
for x in range(0,len(comment_list)):
    for y in range(0,len(comment_list[x])):
        comment_list[x][y] = comment_list[x][y].replace("&amp;", " & ")
        comment_list[x][y] = comment_list[x][y].replace("&lt;", " < ")
        comment_list[x][y] = comment_list[x][y].replace("&quot;", "'")
        comment_list[x][y] = comment_list[x][y].replace("<b>", "")
        comment_list[x][y] = comment_list[x][y].replace("<br>", "")
        comment_list[x][y] = comment_list[x][y].replace("</b>", "")
        comment_list[x][y] = comment_list[x][y].replace("@", "")
        comment_list[x][y] = comment_list[x][y].replace("\n", "")
comment_list[0]
```

```python
for i in range(0,len(comment_list)):
    new_df = pd.DataFrame(comment_list[i])
```

```python
for x in range(0,len(comment_list)):
    for y in range(0,len(comment_list[x])):
        values =comment_list[x][y]
values
```

```python
for x in range(0,len(comment_list)):
    for y in range(0,len(comment_list[x])):
        if 'https' in comment_list[x][y]:
            print(comment_list[x][y])
```

```python
new_df['comments'] = new_df[0] 
new_df = new_df.drop(columns = 0)
```

```python
new_df.head()
```

```python
# Drop duplicates from text column
df = new_df.drop_duplicates(subset = ['Text'],keep='first')
print(df)
```

```python
# Droping rows with href URL
df = df[df["Text"].str.contains(r'href=\S+') == False]
df = df[df["Text"].str.contains(r'https:\S+') == False]
print(df)
```

```python
# Remove or replacing API symbols with originals chained
df.Text = df.Text.str.replace("&#39;", " ' ")\
.str.replace("&amp;", " & ")\
.str.replace("&lt;", " < ")\
.str.replace("&quot;", " ' ")\
.str.replace("<b>", " ")\
.str.replace("<br>", " ")\
.str.replace("</b>", " ")\
.str.replace("@", " ")
print(df)
```

```python
# Using Google translator to translate non English Language to English
result = []

for value in df["Text"]:
    try:
        result.append(translator.translate(value).text)
    except:
        pass
```

```python
# Updating the Text Column with the Translated result
df=pd.DataFrame(result,columns=["Text"])
print(df)
```
